```python
user_input='Krishna Swamy Jalla is the CoFounder of Surge Classes'
user_input
```

```python
import os
import openai

openai.api_key = '' #generate your api_key

response = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
    {
      "role": "system",
      "content": f"Entities extracted from the text \n\ntext: Delhi is the capital of India\n\nentities:\nDelhi: City\nIndia: Country\n\n##\n\ntext: {user_input}"
    }
  ],
  temperature=1,
  max_tokens=256,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)
```
