```python
import cv2
import pytesseract
from google.colab.patches import cv2_imshow
```

```python
# Load the image
img_path = '/content/notes5.jpeg'
img = cv2.imread(img_path, 0)  # Read the image in grayscale

# Preprocess the image
#img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 3, 15)
cv2_imshow(img)
```

```python
pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'
pytesseract.pytesseract.get_languages(config='')[0]  # Download language data
```

```python
text = pytesseract.image_to_string(img)
print(text)
```
