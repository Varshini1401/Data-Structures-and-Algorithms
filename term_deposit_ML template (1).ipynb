```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df_cat['job_cat']=encoder.fit_transform(df_cat['job'])
```

```python
df_cat['marital_cat']=encoder.fit_transform(df_cat['marital'])
df_cat['education_cat']=encoder.fit_transform(df_cat['education'])
df_cat['default_cat']=encoder.fit_transform(df_cat['default'])
df_cat['housing_cat']=encoder.fit_transform(df_cat['housing'])
df_cat['loan_cat']=encoder.fit_transform(df_cat['loan'])
df_cat['month_cat']=encoder.fit_transform(df_cat['month'])
df_cat['poutcome_cat']=encoder.fit_transform(df_cat['poutcome'])
df_cat['y_cat']=encoder.fit_transform(df_cat['y'])
```

```python
df_cat=df_cat.drop(['job','marital','education','default','housing','loan','month','poutcome','y'],axis=1)
```

```python
total_data=pd.concat([df_num,df_cat],axis=1)
total_data.head()
```

```python
total=total_data['y_cat']
```

```python
total_data=total_data.drop(['y_cat'],axis=1)
total_data.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(total_data,total,test_size=0.25,random_state=42)
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(random_state=42)

rfe_RF=RFE(RF,n_features_to_select=4,verbose=21)

rfe_RF.fit(x_train,y_train)
names=x_train.columns.tolist()
b=rfe_RF.ranking_
a=pd.DataFrame(sorted(list(map(lambda x,y: (x,y),b,names))),columns=['rank','features'])
a.to_csv('feature_selection.csv')

RFE_selected=a['features'][a['rank']==1]
```

```python
RFE_selected
```

```python
x_train=x_train[RFE_selected]
```

```python
x_test=x_test[RFE_selected]
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
RF_model=RandomForestClassifier(max_depth=3, max_features=0.6, min_samples_leaf=4,n_estimators=9)
RF_model.fit(x_train,y_train)
```

```python
RF_model
```

```python
pred=RF_model.predict(x_test)
```

```python
out_data=pd.DataFrame({'true':y_test,'model_predicted':pred})
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {"n_estimators" : [90, 180, 270, 360, 450, 540, 630],
              "max_depth" : [2,3,5,7],
              "min_samples_leaf" : [2, 4],
              "max_features" : [0.4,0.5,0.6]
             }

#Use the GridSearchCV
RF_model=RandomForestClassifier() #estimator
clf = GridSearchCV(RF_model, param_grid, cv=3)
clf.fit(x_train,y_train)
clf.best_estimator_
```

```python
clf.best_estimator_
```

```python
preds=clf.best_estimator_.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_test=accuracy_score(y_test,preds)
```
